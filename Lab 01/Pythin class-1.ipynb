## Loops

```python
for i in range(1,11,2):  #(starting,ending, steps difference)
  print(i)
```

```python
for i in range(10,0,-2):  #(starting,ending, steps difference)
  if i%3==0:
    continue
  print(i,end=" ") #end means print in single line
```

```python
i=5
while i>= 0:
  print(i)
  i-=1
```

```python
i=5
while True: #True means infinite
  print(i)
  i-=1
  if i<0:
    break
```

# List



```python
# array - homogenous structure
# list - heterogenous structure
lst= list()
print(type(lst))
lst=[10,20,30,40,60]
print(len(lst))
lst[0]
print(lst[len(lst)-1])
print(lst[-2]) # minus value means printed from last
```

```python
marks=[]
marks.append(78) #append adds an element at the end of the list
print(marks)
```

```python
# marks.append(80) #append adds an element at the end of the list
# marks.insert(1,'AI') #adds an element at the specified position
#marks.pop() #pop removes an element at the end of the list
# marks.pop(1) ##removes an element at the specified position

marks.remove(90)
marks
```

```python
# searching
target=80
if target in marks:
  marks.remove(target)
print(marks)
```

```python
lst=[10,20,30,40,60,100,200,500]
# Iteration: index-based
for i in range(len(lst)):
  print(lst[i])


# Iteration: element-based
for i in lst:
  print(i)
```

# List Slicing

```python
newlist = lst[::2]
print(newlist)
```
